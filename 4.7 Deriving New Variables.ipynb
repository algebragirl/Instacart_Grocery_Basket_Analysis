## 2. If-Statements with User-Defined Functions (with subset)

### 2.1 Define price label function to determine price range in subset

```python
def price_label(row):

    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range product'
    else: return 'Not enough data'
```

### 2.2 Apply the price label function on subset

```python
df['price_range'] = df.apply(price_label, axis=1)
```

### 2.3 Check values in new 'price_range' column in subset

```python
df['price_range'].value_counts(dropna = False)
```

### 2.4 How much is the most expensive product within the subset?

```python
df['prices'].max()
```

## 3. If-Statements with the loc() Function (applied to subset)

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[df['prices'] > 5 & (df['prices'] <= 15), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```
