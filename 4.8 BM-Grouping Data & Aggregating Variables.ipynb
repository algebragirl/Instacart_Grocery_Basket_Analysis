### Task #4. Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
#create a flag that assigns a "loyalty" label to a user ID based on its corresponding max order value.
```

```python
#First, Create a new column 'max_order' 
```

```python
#1. Split the data into groups (use groupby function) based on the 'user_id' column:     groupby(['user_id'])
#2. Apply the transform() function on the 'order_number' column to generate the maximum orders for each user:  ['order_number'].transform('max')
#3. Create a new column, 'max_order,' into which you'll place the results of your aggregation
```

```python
df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
#Second, create a 2nd column, 'loyalty_flag', based on the customer's max orders
# To create the flag, we'll need some criteria:
# 1. If the maximum orders the user has made is over 40, then the customer will be labeled a 'Loyal customer'
# 2. If the maximum orders the user has made is over 10 but less than or equal to 40, then the customer will be labeled a 'Regular customer'
# 3. If the maximum orders the user has made is less than or equal to 10, then the customer will be labeled a 'New customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40 , 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10),'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#check output
df_ords_prods_merge.head()
```

### Task #6. The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying.
### Create a spending flag for each user based on the average price across all their orders using the following criteria:
### - If the mean of the prices of products purchased by a user is lower than 10, then flag them as a "low spender."
### - If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a "High spender."

```python
#First, Create a new column -'average spent' - which applies to each customer
```

```python
# 1. Split data into groups based on the user_id column:  df_ords_prods_merge.groupby(['user_id'])
# 2. Apply the transform() function on the 'prices' column to generate mean price across all their orders:   ['prices'].transform(np.mean)
# 3. Create a new column, 'avg_spent' into which you will place results of your aggregation:  df_ords_prods_merge['avg_spent']
```

```python
df_ords_prods_merge['avg_spent'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#Second, create a 2nd column, 'spending_flag', based on the customer's 'avg_spent'
# To create the flag, we'll need some criteria:
# 1. If the mean of the prices of products purchased by a user is lower than 10, then flag them as a "low spender."
# 2. If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a "High spender."
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['avg_spent'] < 10 , 'spending_flag'] = 'Low spender'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['avg_spent'] >= 10 , 'spending_flag'] = 'High spender'
```

```python
#check output
df_ords_prods_merge.head()
```

```python
df_ords_prods_merge['spending_flag'].value_counts(dropna = False)   #check number and type of spending customers
```

### Task #7. In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again),
### the Instacart team wants you to determine frequent versus non-frequent customers.
### Create an order frequency flag that marks the regularity of a user's ordering behavior
### according to the median in the 'days_since_prior_order' column. The criteria for the flag should be as follows:
### 1. If the median of 'days_since_prior_order' is higher than 20, then the customer should be labeled a "Non-frequent customer."
### 2. If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a "Regular customer."
### 3. If the median of 'days_since_prior_order' is lower than or equal to 10, then the customer should be labeled a "Frequent customer."

```python
#First, Create a new column -'buy_frequency' - which applies to each customer
```

```python
# 1. Split data into groups based on the user_id column:  df_ords_prods_merge.groupby(['user_id'])
# 2. Apply transform() fnct on the 'days_since_prior_order' column to generate mean across all their orders: ['days_since_prior_order'].transform(np.mean)
# 3. Create a new column, 'buy_frequency' into which you will place results of your aggregation:  df_ords_prods_merge['buy_frequency']
```

```python
df_ords_prods_merge['buy_frequency'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.mean)
```

```python
#Second, create a 2nd column, 'frequency_flag', based on the customer's 'buy_frequency'
# To create the flag, we'll need some criteria:
### 1. If the median of 'days_since_prior_order' is higher than 20, then the customer should be labeled a "Non-frequent customer."
### 2. If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a "Regular customer."
### 3. If the median of 'days_since_prior_order' is lower than or equal to 10, then the customer should be labeled a "Frequent customer."
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['buy_frequency'] > 20 , 'frequency_flag'] = 'Non-frequent customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['buy_frequency'] > 10) & (df_ords_prods_merge['buy_frequency'] <= 20) , 'frequency_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['buy_frequency'] <= 10 , 'frequency_flag'] = 'Frequent customer'
```

```python
#check output
df_ords_prods_merge.head()
```

```python
#check number and type of buying customers
df_ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```
